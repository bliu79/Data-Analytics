# Task 4: Understand NY COVID 19 data in last 30 days

```python
dfny= df[df['state']=='New York']
dfny
```

```python
dfny1 = dfny.groupby(['date','state',]) ['cases','deaths'].sum()
dfny1.tail(30)
```

```python
dfny1.tail(30).describe()
```

```python
# plot the case growth curve for NY last 30 days
sns.lineplot (x= 'date', y='cases', data = dfny1.tail(30))
```

```python
# date for NY death rate last 30 days
dfny1_deathrate = dfny1['deaths']/dfny1['cases']
dfny1['death_rate'] = dfny1_deathrate
dfny1.tail(30)
```

```python
# plot death curve for NY last 30 days
sns.lineplot (x= 'date', y='deaths', data = dfny1.tail(30))
```

```python
# Plot death rate curve for NY last 30 days
sns.lineplot (x= 'date', y='death_rate', data = dfny1.tail(30))
```

Understanding of the data:

From the data of NY for the last 30 days, both of cases and deaths are keeping growing rapidly because their growth curves are steeply upward in a similar way. 
However, the death rate keep dropping a lot.

# Task 5: Comparing COVID 19 data from different states

```python
# data for five states
df3=df.query('state in ["New York","Florida","Texas","Illinois","California"]')
df3_deathrate = df3['deaths']/df3['cases']
df3['death_rate'] = df3_deathrate
df3
```

```python
# total cases and deaths
df_5states_groupby= df3.groupby(['date','state'])['cases','deaths'].sum()
df_5states_groupby.tail(10)
```

```python
# adding death rate data for five states
df_5states_deathrate = df_5states_groupby['deaths']/df_5states_groupby['cases']
df_5states_groupby['death_rate'] = df_5states_deathrate
df_5states_groupby.tail(10)
```

```python
df_5states_groupby.describe()
```

```python
# plot case growth curve for five states
sns.lineplot(x='date', y='cases', hue='state', data=df_5states_groupby)
```

```python
# plot death growth curve for five states
sns.lineplot(x='date', y='deaths', hue='state', data=df_5states_groupby)
```

```python
# plot death rate growth curve for five states
sns.lineplot(x='date', y='death_rate', hue='state', data=df_5states_groupby)
```
